```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns',False)
sns.set(rc={'figure.figsize':(20.7,8.27)})
sns.set_style("whitegrid")
sns.color_palette("dark")
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv('data_bersih_ver3.csv')
df['kamar'] = df['kamar_mandi'] + df['kamar_tidur']
df['km:k'] = df['kamar_mandi'] / df['kamar']
df['lb:lt'] =  df['luas_bangunan'] / df['luas_tanah']
df = df[df['tipe']=='Rumah']
df = df.drop(
    ['Unnamed: 0', 'tipe','kamar_mandi','luas_bangunan','taman','garasi','kamar_tidur','jumlah_lantai'], 
    axis = 1
)
df.drop_duplicates()
```

```python
df_copy = df.copy()
```

```python
cat = [i for i in df_copy.columns if df_copy[i].dtypes in ['O','bool']]
df_copy[cat] = df_copy[cat].astype('O')
```

```python
from feature_engine.encoding import DecisionTreeEncoder as DTE
```

```python
encode = DTE(variables = ['kecamatan'])
encode.fit(df_copy,df_copy['harga'])
```

```python
df_copy = encode.transform(df_copy)
df_copy.info()
```

```python
from feature_engine.encoding import OrdinalEncoder as OE
```

```python
oe = OE(variables = [i for i in cat if i != 'kecamatan'])
oe.fit(df_copy,df_copy['harga'])
```

```python
df_copy = oe.transform(df_copy)
```

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python
scale = ss()
scale.fit(df_copy)
df_copy = scale.transform(df_copy)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(
    model,
    k = (2,5),
    metric = 'calinski_harabasz',
    timing = False
)
```

```python
visualizer.fit(df_copy)
visualizer.show()
```

```python
model = KMeans(n_clusters=2)
cluster = model.fit_predict(df_copy)
```

```python
df_copy = pd.DataFrame(df_copy)
df_copy.columns = df.columns
```

```python
df_copy['cluster'] = cluster
```

```python
plt.figure(figsize=(5,2), dpi=200)
df_copy.corr()['cluster'][:-1].abs().sort_values().plot(kind='bar')
```
